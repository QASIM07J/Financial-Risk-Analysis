```python
marital_status_counts = df['Marital Status'].value_counts()
print(marital_status_counts)
```

```python


# Create the bar chart
plt.figure(figsize=(8, 6))
marital_status_counts.plot(kind='bar', color=['blue', 'green', 'red', 'orange'])
plt.title(' Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  

plt.show()
```

```python
# Filter for Single people
single_people = df[df['Marital Status'] == 'Single']

# Count the occurrences of each country for single people
country_counts = single_people['Country'].value_counts()

# Find the country with the most single people
country_with_most_singles = country_counts.index[0]

print("Country with the most single people:", country_with_most_singles)



# Filter for divorced people
divorced_people = df[df['Marital Status'] == 'Divorced']

# Count the occurrences of each country for Divorced people
country_counts = divorced_people['Country'].value_counts()

# Find the country with the most divorced people
country_with_most_divorced = country_counts.index[0]

print("Country with the most divorced people:", country_with_most_divorced)
```

```python
average_loan_amount = df['Loan Amount'].mean()
print(average_loan_amount)
```

```python
danger = 0
for index, row in df.iterrows():
    if row['Loan Amount'] > average_loan_amount and row['Number of Dependents'] > 3:
        danger = danger+1

print(danger)

#It means they have an above average loan and more number of dependents, which is an alarming situation most of the times
```

```python
# Group by Marital Status and calculate total Asset Value
grouped_data = df.groupby('Marital Status')['Assets Value'].sum()

# Find the Marital Status with the highest Asset Value
marital_status_with_highest_asset = grouped_data.idxmax()

print("Marital Status with highest Assets Value:", marital_status_with_highest_asset)
```

```python
plt.figure(figsize=(8, 6))
grouped_data.plot(kind='bar')
plt.title('Total Asset Value by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Total Asset Value')
plt.xticks(rotation=45)

# Format y-axis ticks to display actual values
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: f'{int(x):,}'))

plt.show()
```

```python
#How does loan amount vary by loan purpose?
# Group by Loan Purpose and calculate mean Loan Amount
grouped_data = df.groupby('Loan Purpose')['Loan Amount'].mean()

# Create the bar chart
plt.figure(figsize=(10, 6))
grouped_data.plot(kind='bar')
plt.title('Average Loan Amount by Loan Purpose')
plt.xlabel('Loan Purpose')
plt.ylabel('Average Loan Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
#Age group with their loan amounts (Average amount with each age)

grouped_data = df.groupby('Age')['Loan Amount'].mean()

# Create the bar chart
plt.figure(figsize=(10, 6))
grouped_data.plot(kind='bar')
plt.title('Average Loan Amount by Age')
plt.xlabel('Age')
plt.ylabel('Average Loan Amount')
plt.xticks(rotation=55)
plt.show()
```

```python
#Correlation between variables
# Convert Education Level to numerical values (for correlation calculation like BS = 0, MS = 1 ..)
df['Education Level'] = df['Education Level'].astype('category').cat.codes

# Calculate correlation coefficient
correlation = df['Education Level'].corr(df['Income'])
print("Correlation between Education Level and Income:", correlation)
```

```python
#Value shows there is nearly no relation but we can vizualize using boxplot
sns.boxplot(x='Education Level', y='Income', data=df)
plt.title('Income Distribution by Education Level')
plt.show()
```

```python

# Calculate Debt-to-Income Ratio
df['Debt-to-Income Ratio'] = df['Loan Amount'] / df['Income'] * 100

# Create a box plot
sns.boxplot(x='Marital Status', y='Debt-to-Income Ratio', data=df)
plt.title('Debt-to-Income Ratio by Marital Status')
plt.show()
```

```python
# Create a contingency table for Analysing between Employment status and Education Level.
contingency_table = pd.crosstab(df['Employment Status'], df['Education Level'])

# Visualize using a heatmap
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='Blues')
plt.title('Employment Status vs Education Level')
plt.show()
```

```python
contingency_table = pd.crosstab(df['Gender'], df['Payment History'])

# Visualize using a heatmap
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='Blues')
plt.title('Payment History by Gender')
plt.show()
```

```python
# Create a contingency table
contingency_table = pd.crosstab(df['Gender'], df['Risk Rating'])

# Visualize using a heatmap
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='Blues')
plt.title('Risk Rating by Gender')
plt.show()
```

```python
#The most common loan purpose
# Count the occurrences of each loan purpose
loan_purpose_counts = df['Loan Purpose'].value_counts()

# Get the most common loan purpose
most_common_loan_purpose = loan_purpose_counts.index[0]

print("The most common loan purpose is:", most_common_loan_purpose)
```

```python
# Calculate Debt-to-Income Ratio
df['Debt-to-Income Ratio'] = (df['Loan Amount'] / df['Income']) * 100

# Visualize the relationship
sns.scatterplot(x='Income', y='Debt-to-Income Ratio', data=df)
plt.title('Income vs Debt-to-Income Ratio')
plt.show()
```

```python
# Create a cross tabulation
ct = pd.crosstab(df['Marital Status'], df['Loan Purpose'])

# Create the grouped bar chart
ct.plot(kind='bar', stacked=False)
plt.title('Loan Purpose by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
